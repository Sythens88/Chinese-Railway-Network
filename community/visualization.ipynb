```python
#pip install geoplot
#pip install geopandas
import geoplot as gplt
import geopandas as gpd
import geoplot.crs as gcrs
import pandas as pd
import networkx as nx
import random
import community
```

```python
df = pd.read_csv('com.csv')
df.columns = ['station','lng','lat','com']
#df['com'] = df['com'].apply(lambda x:'community'+str(x))
df = df.set_index('station')
```

```python
df['size'] = [random.random() for _ in range(len(df))]
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lng, df.lat))
```

```python
china = gpd.read_file("chinageo.shp")
ax = gplt.polyplot(china)
gplt.pointplot(gdf,ax=ax,hue='com',cmap='rainbow',legend=False,scale='size',limits=(2,3))
```
